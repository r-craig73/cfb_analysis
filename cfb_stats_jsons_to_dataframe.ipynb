Import Python modules.

```python
import json
import pandas as pd
import os
```

Load regular and postseason games JSON from a year into a pandas dataframes.

```python
cfb_year = input('Input college football year?')

json_regular_season = cfb_year + '_cfb_JSON-CSV/' + cfb_year + '-basicGameStatsCFB-JSON-regular.json'
json_post_season = cfb_year + '_cfb_JSON-CSV/' + cfb_year + '-basicGameStatsCFB-JSON-postseason.json'

with open(json_regular_season) as jrs:
    regular_season_games = json.load(jrs)

with open(json_post_season) as jps:
    postseason_games = json.load(jps)

# # DataFrames from dict of array-like or dicts
df01 = pd.DataFrame.from_dict(regular_season_games)
df02 = pd.DataFrame.from_dict(postseason_games)

# New DataFrame, but selecting 11 different columns, not 33 columns
regular_season_games = pd.DataFrame(df01, columns=['id', 'season', 'season_type', 'week', 'completed', 'home_team', 'home_division', 'home_points', 'away_team', 'away_division', 'away_points'])
postseason_games = pd.DataFrame(df02, columns=['id', 'season', 'season_type', 'week', 'completed', 'home_team', 'home_division', 'home_points', 'away_team', 'away_division', 'away_points'])

# Concatenate regular_season_games DataFrame and postseason_games DataFrame
cfb_season_games = pd.concat([regular_season_games, postseason_games], axis=0)
```
