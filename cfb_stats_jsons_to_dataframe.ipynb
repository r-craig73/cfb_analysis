Import Python modules.

```python
import json
import pandas as pd
import os
```

Load games JSON from a year into a pandas dataframe.

```python
cfb_year = input('Input college football year?')

json_regular_season = cfb_year + '_cfb_JSON-CSV/' + cfb_year + '-basicGameStatsCFB-JSON-regular.json'

with open(json_regular_season) as fp:
    games = json.load(fp)

# DataFrame from dict of array-like or dicts
df = pd.DataFrame.from_dict(games)

# New DataFrame, but selecting 10 different columns, not 33 columns
df2 = pd.DataFrame(df, columns=['id', 'season', 'week', 'completed', 'home_team', 'home_division', 'home_points', 'away_team', 'away_division', 'away_points'])
df2
```
